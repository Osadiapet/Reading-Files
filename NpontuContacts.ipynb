```python
#import required libraries which contains
import pandas as pd 
import numpy as np
```

```python
#loading the raw csv data file 
data=pd.read_csv('contacts.csv')
data
```

```python
##removing spaces from the dataset
data['new_contact']=data["Contact"].str.replace(" ", "")
data['new_contact']  #all spaces cleared
```

```python
#prefixing first index with 233 in order to get a uniform data
if data['new_contact'].all().startswith('0'):
    data['new_contact']=data['new_contact'].str.replace(data['new_contact'].all()[0],'233')
```

```python
data['new_contact']
```

```python
##deleting the contact column sice we dont need it again
data=data.drop(['Contact'], axis=1)
data
```

```python
#splitting the data into two (60% and 40%)
from sklearn.model_selection import train_test_split
urlA, urlB=train_test_split(data, test_size=0.4)
```
